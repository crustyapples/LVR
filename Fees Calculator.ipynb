```python
import requests
from datetime import datetime

# Define the GraphQL endpoint
endpoint = 'https://api.thegraph.com/subgraphs/name/uniswap/uniswap-v3'

# Define the GraphQL query
query = '''
query poolDayDatas($startTime: Int!,$endTime: Int!, $skip: Int!, $address: Bytes!) {
  poolDayDatas(
    first: 1000
    skip: $skip
    where: {pool: $address, date_gt: $startTime}
    orderBy: date
    orderDirection: asc
    subgraphError: allow
  ) {
    date
    volumeUSD
    tvlUSD
    feesUSD
    pool {
      feeTier
      __typename
    }
    __typename
  }
}
'''

# Get the current timestamp and the desired timeframe
start_timestamp = 1684395515
end_timestamp = 1688543783

# Set the query variables
variables = {
    "startTime": start_timestamp,
    "endTime": end_timestamp,
    "skip": 0,
    "address": "0x88e6a0c2ddd26feeb64f039a2c41296fcb3f5640"
}

# Define the GraphQL request payload
payload = {
    "query": query,
    "variables": variables
}

# Send the GraphQL request
response = requests.post(endpoint, json=payload)

# Process the query response
fees = 0
if response.status_code == 200:
    data = response.json().get('data')
    if data:
        poolDayDatas = data.get('poolDayDatas')
        for dayData in poolDayDatas:
            # check that date is within the desired timeframe
            if dayData['date'] >= start_timestamp and dayData['date'] <= end_timestamp:
              print(datetime.fromtimestamp(dayData['date']), dayData['feesUSD'])
              fees += float(dayData['feesUSD'])

# Display the total fees collected from start date to end date using the timestamps
print(f'Total fees collected from {datetime.fromtimestamp(start_timestamp)} to {datetime.fromtimestamp(end_timestamp)}: {fees}')
```
